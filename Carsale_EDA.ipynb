## Predicting Future Sales by Machine Learning

```python
#Splitting dataset based on dependent and independent feature

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#Label-Encoding to better fit the model
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()  #creating instance of LabelEncoder
data['Model_No'] = labelencoder.fit_transform(data['Model'])
#land_rover_sales['Model_No'] = labelencoder.fit_transform(land_rover_sales['Model']).  #For LandRover Models

x = data.drop(['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec','Year','Model', 'Car_Size', 'Total_Sales'], axis=1).iloc[:, :].values    # independent feature
y = data.Total_Sales  # dependent feature

# Splitting the dataset into the Training set and Test set
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=1)

#Regression Object
lr = LinearRegression()
lr.fit(x_train,y_train)

plt.scatter(x_train[:,0],y_train,color = 'red')
plt.plot(x_train,lr.predict(x_train),color = 'green')
plt.title("Model Vs. Sales ( TRAINING SET)")
plt.xlabel("Model")
plt.ylabel("Total Sales")
plt.grid()
plt.show()
```

```python
#Plotting the results
plt.scatter(x_test,y_test,color = 'red')
plt.plot(x_test,lr.predict(x_test),color = 'black')
plt.title("Model Vs. Sales (TEST SET)")
plt.xlabel("Model")
plt.ylabel("Total Sales")
plt.grid()
plt.show()
```

```python
# ACTUAL Vs PREDICTED Total Sale
y_pred=lr.predict(x_test)
df2=pd.DataFrame({'Actual Sale':y_test,'Predicted Sale':y_pred})
df2
```
